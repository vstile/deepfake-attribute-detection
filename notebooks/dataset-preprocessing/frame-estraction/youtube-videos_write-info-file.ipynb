```python
import os

# Percorsi delle cartelle
faces_dir = "/Volumes/HDD2 - 12Tb/Developer/GitHub/PhD-Projects/dataset/dataset/manipulated_sequences/Deepfakes/c40/faces"
frames_dir = "/Volumes/HDD2 - 12Tb/Developer/GitHub/PhD-Projects/dataset/dataset/manipulated_sequences/Deepfakes/c40/frames"
info_txt_path = "/Volumes/HDD2 - 12Tb/Developer/GitHub/PhD-Projects/dataset/dataset/manipulated_sequences/Deepfakes/c40/info.txt"

# Funzione per contare i file che iniziano con un prefisso specifico in una cartella
def count_files_with_prefix(directory, prefix):
    if not os.path.exists(directory):
        return 0
    return len([name for name in os.listdir(directory) if name.startswith(prefix) and os.path.isfile(os.path.join(directory, name))])
```

Use this, instead, to count and write only faces:

```python
# Creare il file di info.txt con conteggio faces per video da 000.mp4 a 999.mp4
print("Inizio creazione file info.txt")
with open(info_txt_path, 'w') as info_file:
    for i in range(1000):
        video_name = f"{i:03}.mp4"  # Generare il nome del video con tre cifre
        prefix = f"{i:03}"  # Estrarre le prime tre cifre del nome del video
        num_faces = count_files_with_prefix(faces_dir, prefix)
        info_file.write(f"Video: {video_name}, estratti {num_faces} volti\n")
        info_file.flush()  # Assicurarsi che i dati vengano scritti nel file
        print(f"Scrittura info per {video_name}: {num_faces} volti")
print("Terminata creazione file info.txt")
```
