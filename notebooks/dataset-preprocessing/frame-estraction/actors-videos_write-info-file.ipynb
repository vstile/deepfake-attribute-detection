```python
import os

# Percorsi delle cartelle
videos_dir = "/Volumes/Macintosh SSD/Developer/FaceForensics++/original_sequences/actors/c40/videos"
frames_dir = "/Volumes/Macintosh SSD/Developer/FaceForensics++/original_sequences/actors/c40/frames"
faces_dir = "/Volumes/Macintosh SSD/Developer/FaceForensics++/original_sequences/actors/c40/faces"
info_txt_path = "/Volumes/Macintosh SSD/Developer/FaceForensics++/original_sequences/actors/c40/info.txt"

# Funzione per contare i file che contengono un prefisso specifico in una cartella
def count_files_with_prefix(directory, prefix):
    if not os.path.exists(directory):
        print(f"Directory non trovata: {directory}")
        return 0
    print(f"Contando i file in {directory} con prefisso {prefix}")
    return len([name for name in os.listdir(directory) if prefix in name and os.path.isfile(os.path.join(directory, name))])

# Creare il file di info.txt
print("Inizio creazione file info.txt")
with open(info_txt_path, 'w') as info_file:
    # Ordina i file in videos_dir
    print(f"Elencando i file in {videos_dir}")
    video_files = sorted([f for f in os.listdir(videos_dir) if os.path.isfile(os.path.join(videos_dir, f))])
    print(f"Trovati {len(video_files)} file video")
    
    for video_file in video_files:
        print(f"Elaborando il file video: {video_file}")
        video_name = video_file  # Nome del video
        prefix = os.path.splitext(video_file)[0]  # Nome del file senza estensione
        num_frames = count_files_with_prefix(frames_dir, prefix)
        num_faces = count_files_with_prefix(faces_dir, prefix)
        info_file.write(f"Video: {video_name}, {num_frames} frame, estratti {num_faces} volti\n")
        info_file.flush()  # Assicurarsi che i dati vengano scritti nel file
        print(f"Scrittura info per {video_name}: {num_frames} frame, {num_faces} volti")
print("Terminata creazione file info.txt")
```
