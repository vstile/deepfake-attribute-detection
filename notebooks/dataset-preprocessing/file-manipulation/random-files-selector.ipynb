```python
import os
import shutil
import random

# Percorsi delle cartelle
FAKE_FACES_PATH = "/Volumes/Macintosh SSD/FaceForensics_c40/manipulated_sequences/Deepfakes/c40/faces"
REAL_FACES_PATH = "/Volumes/Macintosh SSD/FaceForensics_c40/original_sequences/youtube/c40/faces"
DEST_FAKE_PATH = "/Volumes/Macintosh SSD/Developer/TEST 4000 faces/FAKE_faces_4000"
DEST_REAL_PATH = "/Volumes/Macintosh SSD/Developer/TEST 4000 faces/REAL_faces_4000"

# Creazione delle cartelle di destinazione se non esistono
os.makedirs(DEST_FAKE_PATH, exist_ok=True)
os.makedirs(DEST_REAL_PATH, exist_ok=True)

# Funzione per copiare file casuali da una cartella sorgente a una cartella di destinazione
def copy_random_files(src_path, dest_path, num_files=4000):
    all_files = os.listdir(src_path)
    random_files = random.sample(all_files, num_files)
    for file_name in random_files:
        full_file_name = os.path.join(src_path, file_name)
        if os.path.isfile(full_file_name):
            shutil.copy(full_file_name, dest_path)

# Copia dei file
copy_random_files(FAKE_FACES_PATH, DEST_FAKE_PATH)
copy_random_files(REAL_FACES_PATH, DEST_REAL_PATH)
```
